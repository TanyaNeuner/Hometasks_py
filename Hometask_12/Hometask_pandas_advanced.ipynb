### Задание 1

```python
def type_of_traffic(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        return 'organic'
    elif row['traffic_source'] == 'paid' or row['traffic_source'] == 'email':
        if row['region'] == 'Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return row['traffic_source']
        
log = pd.read_csv('visit_log.csv', sep = ';')
log['source_type'] = log.apply(type_of_traffic, axis = 1)
log.head(10)
```

### Задание 3

```python
ratings = pd.read_csv('ratings.csv')
ratings['timestamp_converted'] = pd.to_datetime(ratings['timestamp'], unit = 's')
# ratings.head(10)

grouped_ratings = ratings.groupby('userId').agg({'timestamp_converted': ['min', 'max'],
                                                          'rating': 'count'})
grouped_ratings.loc[grouped_ratings['rating']['count'] > 100, 'lifetime'] = grouped_ratings['timestamp_converted']['max'] - grouped_ratings['timestamp_converted']['min']
filtered_grouped_ratings = grouped_ratings[~pd.isna(grouped_ratings['lifetime'])]
filtered_grouped_ratings.head(10)
```

### К домашнему заданию, задача 2
Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
rzd['category'] = 'РЖД'
auto['category'] = 'Автомобиль'
air['category'] = 'Авиа'

def revenue(row):
    if not pd.isna(row['rzd_revenue']):
        return row['rzd_revenue']
    elif not pd.isna(row['auto_revenue']):
        return row['auto_revenue']
    elif not pd.isna(row['air_revenue']):
        return row['air_revenue']
    
revenue_summary = pd.concat([rzd,auto,air], axis = 0)
revenue_summary['revenue'] = revenue_summary.apply(revenue, axis = 1)
```

##### Задание 4.1 (без указания адреса)

```python
revenue_summary.pivot_table(index = 'client_id',columns = 'category', values = 'revenue', aggfunc = 'sum', fill_value = 0, margins = False)
```

##### Задание 4.2 (с указанием адреса)

```python
pt_revenue = revenue_summary.pivot_table(index = 'client_id',columns = 'category', values = 'revenue', aggfunc = 'sum', fill_value = 0, margins = False).reset_index()
merged_data = pt_revenue.merge(client_base, on = 'client_id', how = 'outer')
merged_data.head(50)
```
