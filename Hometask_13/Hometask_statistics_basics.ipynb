### Задание

#### Описание полей
##### surgery?
      1 = Yes, it had surgery
      2 = It was treated without surgery
##### Age
      1 = Adult horse
      9 = Young (< 6 months)
##### rectal temperature
      - linear
      - in degrees celsius.
      - An elevated temp may occur due to infection.
      - temperature may be reduced when the animal is in late shock
      - normal temp is 37.8
      - this parameter will usually change as the problem progresses
           eg. may start out normal, then become elevated because of the lesion, passing back through the normal range as
           the horse goes into shock   
##### pulse
      - linear
      - the heart rate in beats per minute
      - is a reflection of the heart condition: 30 -40 is normal for adults
      - rare to have a lower than normal rate although athletic horses may have a rate of 20-25
      - animals with painful lesions or suffering from circulatory shock may have an elevated heart rate
##### respiratory rate
      - linear
      - normal rate is 8 to 10
      - usefulness is doubtful due to the great fluctuations
##### temperature of extremities
      - a subjective indication of peripheral circulation
      - possible values:
           1 = Normal
           2 = Warm
           3 = Cool
           4 = Cold
      - cool to cold extremities indicate possible shock
      - hot extremities should correlate with an elevated rectal temp.
##### pain 
      - a subjective judgement of the horse's pain level
      - possible values:
           1 = alert, no pain
           2 = depressed
           3 = intermittent mild pain
           4 = intermittent severe pain
           5 = continuous severe pain
      - should NOT be treated as a ordered or discrete variable!
      - In general, the more painful, the more likely it is to require surgery
      - prior treatment of pain may mask the pain level to some extent
##### outcome
      - what eventually happened to the horse?
      - possible values:
           1 = lived
           2 = died
           3 = was euthanized

Поля surgery?, Age, temperature of extremities, pain и outcome содержат *качественные данные*.

Поля rectal temperature, pulse и respiratory rate содержат *количественные непрерывные величины*.


```python
# получение датафрейма
horse_data = pd.read_csv('horse_data.csv', names = ['surgery?', 'Age', 'Hospital Number', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'peripheral pulse', 'mucous membranes', 'capillary refill time', 'pain', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination - feces', 'abdomen', 'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome', 'surgical lesion?', 'site of lesion', 'type of lesion', 'subtype of lesion', 'specific code of lesion', 'cp_data'], na_values = '?')
horse_data.head(10)
```

```python
# фильтрация обозначенных 8 столбцов
filtered_horse_data = horse_data[['surgery?', 'Age','rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']]
filtered_horse_data.head(10)
```

```python
# описательная статистика
filtered_horse_data.describe()
```

```python
print("Мода поля surgery?:", filtered_horse_data['surgery?'].mode()[0])
print("Мода поля Age:", filtered_horse_data['Age'].mode()[0])
print("Мода поля rectal temperature:", filtered_horse_data['rectal temperature'].mode()[0])
print("Мода поля pulse:", filtered_horse_data['pulse'].mode()[0])
print("Мода поля respiratory rate:", filtered_horse_data['respiratory rate'].mode()[0])
print("Мода поля temperature of extremities:", filtered_horse_data['temperature of extremities'].mode()[0])
print("Мода поля pain:", filtered_horse_data['pain'].mode()[0])
print("Мода поля outcome:", filtered_horse_data['outcome'].mode()[0])
```

##### Стандартное отклонение
В среднем разброс величин от средней:
1) 0.73 градуса по ректальной температуре (от 38.17),
2) 28.63 удара/мин по пульсу (от 71.9),
3) 17.64 вдоха/мин по частоте дыхания (от 30.42).



```python
## выбросы
# ректальная температура
q1 = filtered_horse_data['rectal temperature'].quantile(0.25)
q3 = filtered_horse_data['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers_rec_temp = filtered_horse_data[filtered_horse_data['rectal temperature'].between(lower_bound, upper_bound, inclusive = True)]
concatenated_data_rec_temp = pd.concat([filtered_horse_data,remove_outliers_rec_temp])
concatenated_data_rec_temp[~pd.isna(concatenated_data_rec_temp['rectal temperature'])].drop_duplicates(keep = False).sort_values(by = 'rectal temperature')
```

```python
# пульс
q1 = filtered_horse_data['pulse'].quantile(0.25)
q3 = filtered_horse_data['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers_pulse = filtered_horse_data[filtered_horse_data['pulse'].between(lower_bound, upper_bound, inclusive = True)]
concatenated_data_pulse = pd.concat([filtered_horse_data,remove_outliers_pulse])
concatenated_data_pulse[~pd.isna(concatenated_data_pulse['pulse'])].drop_duplicates(keep = False).sort_values(by = 'pulse')
```

```python
# частота дыхания
q1 = filtered_horse_data['respiratory rate'].quantile(0.25)
q3 = filtered_horse_data['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers_resp_rate = filtered_horse_data[filtered_horse_data['respiratory rate'].between(lower_bound, upper_bound, inclusive = True)]
concatenated_data_resp_rate = pd.concat([filtered_horse_data,remove_outliers_resp_rate])
concatenated_data_resp_rate[~pd.isna(concatenated_data_resp_rate['respiratory rate'])].drop_duplicates(keep = False).sort_values(by = 'respiratory rate')
```

#### Обработка пропусков
##### surgery? и outcome
В полях пропущено по одному значению и эти значения соответствуют одному наблюдению. Замена этих пропусков расчетными или аналитическими значениями невозможна, поскольку это исказит данные. Удаление одной строки из 300 в данной ситуации исказит данные незначительно.

##### частота дыхания
Поскольку поле не является значимым для анализа из-за сильной флуктуации, данное поле может быть полностью удалено при наличии в нем пропусков.

##### температура, пульс, температура конечностей, боль
В датасете присутствуют строки, где не заполнена информация по выше перечисленным полям. Данные пропуски не могут быть заполнены средними значениями, поскольку в разрезе данных о хирургическом вмешательстве, возрасте и исходе значения разнятся. Удаление этих строк приведет к потере всего 3% данных, что в принятых пределах достоверности в 95% можно считать допустимым.

##### пульс
По описанию, при болезненном поражении органов и острой сердечной недостаточности может наблюдаться повышенный уровень пульса. Поэтому для более точного определения пропущенных значений пульса взята группировка по боли. 
Также было замечено, что пульс разнится для лошадей разных возрастных групп и в зависимости от хирургического вмешательства, поэтому данные показатели также были учетны.
Пустоты заполнены медианными значениями по группировке: возраст, хирургическое вмешательство, боль.

##### температура
По описанию, повышенная температура наблюдается вследствие инфекции, а пониженная вследствие сердечной недостаточности. Однако отклонения замечены для различных возрастов и исходов для животного.
Пустоты заполнены медианными значениями по группировке: возраст, исход.

##### температура конечностей
По описанию, горячие конечности должны коррелировать с повышенной температурой. На данных это однозначно не подтверждается.
Пустоты заполнены медианными значениями по группировке: исход, температура.
Таким образом заполнилось 31 пустое значение из 48.
Для дозаполнения применено медианное значение по группировке: возраст, исход, боль, хирургическое вмешательство.

##### боль
По описанию, в обще случае, чем выше боль, тем более вероятно, что понадобится хирургическое вмешательство, что частично доказывается данными. Также выявлена некая взаимосвязь с исходом животного и его возрастом.
Заполнение столь субъективных полей расчетными, однако, остается под сомнением.
Пустоты заполнены медианными значениями по группировке: возраст, хирургическое вмешательство, исход.

В результате заполнения пропусков отмечены незначительные изменения в значениях средней, моды и медианы.

```python
# удаление столбца с частотой дыхания
dealing_with_na_horse_data = filtered_horse_data.drop(['respiratory rate'], axis = 'columns')
dealing_with_na_horse_data.info()
```

```python
# удаление пустых строк по данным о хирургическом вмешательстве и исходе
dealing_with_na_horse_data = dealing_with_na_horse_data.dropna(subset = ['surgery?','outcome'])
dealing_with_na_horse_data.info()
```

```python
# удаление строк, где пусты значения температуры, пульса, температуры конечностей и боли
dealing_with_na_horse_data = dealing_with_na_horse_data.dropna(thresh = 1, subset = ['rectal temperature', 'pulse','temperature of extremities','pain'])
dealing_with_na_horse_data.info()
```

```python
# данные для rectal temperature
filtered_horse_data.groupby(['Age','surgery?','pain']).agg('median')
```

```python
# данные для pulse
filtered_horse_data.groupby(['Age','outcome']).agg('median')
```

```python
# данные для temperature of extremities

filtered_horse_data.groupby(['Age','outcome','pain','surgery?']).agg('median')
```

```python
# данные для pain
filtered_horse_data.groupby(['Age','outcome','surgery?']).agg('median')
```

```python
# заполнение пустоты по пульсу
dealing_with_na_horse_data['pulse'].fillna(filtered_horse_data.groupby(['Age','surgery?','pain'])['pulse'].transform('median'), inplace = True)
dealing_with_na_horse_data.info()
```

```python
# заполнение пустоты по температуре
dealing_with_na_horse_data['rectal temperature'].fillna(filtered_horse_data.groupby(['Age','outcome'])['rectal temperature'].transform('median'), inplace = True)
dealing_with_na_horse_data.info()
```

```python
# заполнение пустоты по температуре конечностей
dealing_with_na_horse_data['temperature of extremities'].fillna(filtered_horse_data.groupby(['outcome','rectal temperature'])['temperature of extremities'].transform('median').round(0), inplace = True)
dealing_with_na_horse_data.info()
```

```python
# заполнение пустоты по боли
dealing_with_na_horse_data['pain'].fillna(filtered_horse_data.groupby(['Age','surgery?','outcome'])['pain'].transform('median').round(0), inplace = True)
dealing_with_na_horse_data.info()
```

```python
# заполнение пустоты по температуре конечностей
dealing_with_na_horse_data['temperature of extremities'].fillna(dealing_with_na_horse_data.groupby(['Age','outcome','pain','surgery?'])['temperature of extremities'].transform('median').round(0), inplace = True)
dealing_with_na_horse_data.info()
```
