### Задание 1
Для проверки корректности номера также добавлена проверка верности региона - не через регулярные выражения, а просто через Python

```python
import re

region_list = [str(el) for el in list(range(10,100))]
extra_list_1 = ['102', '702', '113', '116', '716', '121', '122', '123', '193', '124', '125', '126', '127', '134', '136', '138', '142', '147', '150', '190', '750', '790', '152', '154', '156', '159', '161', '761', '163', '763', '164', '196', '169', '173', '174', '774', '177', '199', '197', '777', '797', '799', '178', '198', '186']
extra_list_2 = [str(el) for el in list(range(1,10))]

for el in extra_list_2:
    el = '0' + el
    region_list.append(el)

region_list.extend(extra_list_1)

car_id = input('Введите транспортный номер: ')
pattern_car_id_check = r'[АВЕКМНОРСТУХ]{1}\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}'
if re.findall(pattern_car_id_check, car_id) != []:
    pattern_car_id_true = r'[АВЕКМНОРСТУХ]{1}\d{3}[АВЕКМНОРСТУХ]{2}'
    pattern_region_true = r'\d{2,3}'
    print(f'Номер {re.findall(pattern_car_id_true, car_id)[0]} валиден.')
    if re.findall(pattern_region_true, car_id)[1] in region_list:
        print(f'Регион: {re.findall(pattern_region_true, car_id)[1]}')
    else:
        print(f'Регион {re.findall(pattern_region_true, car_id)[1]} не валиден')
```

### Задание 4

```python
import re

emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
domains_list = []

for email in emails:
    domains_list.extend(re.findall(r'@\w+\.\w+', email))

domains_set = set(domains_list)

for domain in domains_set:
    print(f'{domain}: {domains_list.count(domain)}')
```
